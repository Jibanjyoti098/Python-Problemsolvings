```python
"""
*
  *
    *
"""
n = 3
for i in range(n):
    for j in range(n):
        if i == j:
            print('*', end=' ')
        else:
            print(' ', end=' ')

    print()
```

```python
n = 5
for i in range(n):
    for j in range(n):
        if i+j ==n-1 or i == j:
            print('*', end=' ')
        else:
            print(' ', end=' ')
    print()
```

```python
"""
      *
    *
  *

"""
n = 3
sp = n
st = 1
i = 0
while i<n:
    for j in range(sp):
        print(' ', end=' ')
        sp-=1
    for j in range(st):
        print('*')
    print()
    i+=1
```
