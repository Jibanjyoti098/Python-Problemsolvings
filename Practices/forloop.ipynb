```python
#Reverse string without slicing
s = input('Enter string: ')
ns = ''
for i in range(0,len(s)):
    ns = s[i]+ns
print(ns)
```

```python
#Extract lowercase char in a str only if ascii value is even
s = input('Enter string: ')
lc  = ''
for i in s:
    if 'a'<=i<='z' and ord(i)%2==0:
        lc+=i
print(lc)
```
