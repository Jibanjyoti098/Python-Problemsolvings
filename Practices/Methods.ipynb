```python
class A:
    name = 'Rahul'
    state = 'Odisha'
    dist = 'Puri'
    ps = 'Town PS'

    def __init__(self, no_of_eyes, no_of_nose , no_of_hand , no_of_ear):
        self.eyes=no_of_eyes
        self.nose=no_of_nose
        self.hand=no_of_hand
        self.ear = no_of_ear


    #accessing
    def object_dispaly(self):
        print(f'''My name is {self.name}
no of eyes is {self.eyes}
no of nose is {self.nose}
no of hand is {self.hand}
NO of ear is {self.ear}''')


    #modifying
    def ch_eye(self,new):
        self.n_no_of_eyes=new
    def ch_nose(self,new):
        self.n_no_of_nose=new
    def ch_no_of_hand(self,new):
        self.n_no_of_hand = new
    def ch_no_of_ear(self,new):
        self.n_no_of_ear = new


    #accessing
    @classmethod
    def class_display(cls):
        print(f'''
Name is {cls.name}
state is {cls.state}
dist is {cls.dist}
ps is {cls.ps}''')

    #modification
    @classmethod
    def ch_name(cls,new):
        cls.name = new

    @classmethod
    def ch_state(cls,new):
        cls. state= new

    @classmethod
    def ch_dist(cls,new):
        cls. dist= new

    @classmethod
    def ch_ps(cls,new):
        cls.ps=new

    @staticmethod
    def wlc():
        print('Wel to Qspider')

    @staticmethod
    def print_hello(n):
        print(n * 'Hello')

    @staticmethod
    def even_odd_check(n):
        if n%2==0:
            print('Even')
        else:
            print('odd')

    @staticmethod
    def find_largest(a,b,c):
        if a>b and a>c:
            print(a)
        elif b>a and b>c:
            print(b)
        else:
            print(c)

    @staticmethod
    def find_Sq(n):
        print(n ** 2)


obj = A(2,1,2,2)
obj.object_dispaly()
obj.class_display()

print('Static method start ')
print()
print()
obj.wlc()
obj.print_hello(4)
obj.even_odd_check(6)
obj.find_largest(29,38,10)
obj.find_Sq(7)
```
