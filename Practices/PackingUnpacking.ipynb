# **Packing**

*args

```python
def Packing(*args):
    return args
print(Packing('pack', 'me'))
```

```python
#Sum of numbers using *args
def Sum(*args, s = 0):
    for i in args:
        s+=i
    return s
Sum(1,2)
```

```python
def func(a,b):
    print(a,b)
func(2,3)
def func(*args):
    print(args)
func(2,3)
```

```python
def Powder(name, *args):
    print(name,args)
Powder('risi', 23,34,45)
```

```python
#Max using args
def find_max(*args):
    return max(args)
find_max(12,23,43)
```

```python
# Multiply all nums using *args:
def mul(*args, m=1):
    for i in args:
        m*=i
    return m
mul(2,2,3,4,5,6)
```

```python
def display(*args):
    for item in args:
        print(item)

display("Python", "Java", "C++")
```

**kwargs:

```python
def Packing(**kwargs):
    return kwargs
print(Packing(a='pack', b='me'))
```

```python
def show(**kwargs):
    print(kwargs)
show(name='Rahul', age=25)
```

```python
def show(**kwargs):
    print(kwargs['name'])
    print(kwargs['age'])
show(name='Rahul', age=25)
```

```python
def display(**kwargs):
    for i,j in kwargs.items():
        print(i,":",j)
display(name='risi',city='bbsr',age=23)
```
