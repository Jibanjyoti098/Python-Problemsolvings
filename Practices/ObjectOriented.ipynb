Class and objects

```python
class Car:
    def __init__(self,brand, model):
        self.brand = brand
        self.model = model

my_car = Car('Toyota', 'Corolla')
print(my_car.brand)
print(my_car.model)
```

```python
class Sahil:
    a = 10
    b = 20
o1 = Sahil()
print(o1)
print(Sahil)
print(Sahil())
print(Sahil().a)
```

```python
class Person:
    def __init__(self,name, age):
        self.name  = name
        self.age  = name
    def greet(self):
        print(f'helllo {self.name} you are {self.age} years old')
o1 = Person('Debadatta', 21)
o1.greet()
```
