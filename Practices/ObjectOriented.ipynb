1. Object: A bundle of related attributes and methods.
2. class: Blueprint used to design the structure and layout of an object.

```python
class Car:
    def __init__(self, model, year, color, for_sale):
        self.model = model
        self.year = year
        self.for_sale = for_sale
        self.color = color
    def drive(self):
        print(f'You drive the {self.model}.')

    def stop(self):
        print(f'You stop the {self.model}.')

car1 =  Car('Musatang', 2024, 'red', False)
car2 =  Car('Corvette', 2025, 'blue', True)

print(car1.model)
print(car1.year)
print(car1.color )
print(car1.for_sale)
print()
print(car2.model)
print(car2.year)
print(car2.color )
print(car2.for_sale)
print()
print(car1.drive())
print(car1.stop())
```

Inheritance: Allow a class to inherit attributes and methods from another class and helps with code reusability and extensibility

```python
class Animal:
    def __init__(self, name):
        self.name = name
        self.is_alive = True

    def eat(self):
        print(self.name,"is eating")

    def sleep(self):
        print(self.name,"is sleeping")

class Dog(Animal):
    def speak(self):
        print('woof!')
class Cat(Animal):
    def speak(self):
        print('Meaw')
class Mouse(Animal):
    def speak(self):
        print('SQUEEK')
dog = Dog('Kaalu')
cat = Cat('Sundari')
mouse = Mouse('Stuad')
cat.is_alive = False

print(dog.sleep())
print(dog.name, dog.is_alive)
print(cat.name, cat.is_alive)
print(mouse.name, mouse.is_alive)
```

Multiple c(A,B) and Multilevel c(B)< B(A)< A Inheritance:

```python
class Animal:
    def __init__(self, name):
        self.name =  name
    def eat(self):
        print(f'{self.name} animal is eating')

    def sleep(self):
        print(f'{self.name} animal is sleeping')

class Pray(Animal):
    def flee(self):
        print(f'{self.name} animal is fleeing')
class Predator(Animal):
    def hunt(self):
        print(f'{self.name} is hunting')
class Rabbit(Pray):
    pass
class Hawk(Predator):
    pass
class Fish(Pray, Predator):
    pass

rabit = Rabbit('Thekua')
hawk = Hawk('kalia')
fish = Fish('macha')
rabit.flee()
hawk.hunt()
fish.flee()
fish.hunt()
rabit.eat()
rabit.sleep()
```

1.Abstract Class:A class can not be instantiated on its own; Meant to be sub class
2.Abstract method can be declared but have no implementation.

```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def go(self):
        pass
    @abstractmethod
    def stop(self):
        pass
class Car(Vehicle):
    def go(self):
        print('You drive the car')
    def stop(self):
        print('You stop the car')

class Motorcycle(Vehicle):
    def go(self):
        print('You ride the motorcycle')
    def stop(self):
        print('You stop the motorcycle')

class Boat(Vehicle):
    def go(self):
        print('You sail the boat')

motorcycle = Motorcycle()
motorcycle.go()
```
