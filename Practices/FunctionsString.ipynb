# Functions of String

## Remove spaces from the string:
1. rstrip()
2. lstrip()
3. strip()

```python
from curses.ascii import isalnum

s = '    I am kingraa  '
print(len(s))

s1 = s.rstrip()
print(s1)
print(len(s1))

s2 = s.lstrip()
print(s2)
print(len(s2))

s3 = s.strip()
print(s3)
print(len(s3))
```

# Find Substring:

## There are 2 function to find the substring
1. find()
2. index()

 1.find()

```python
s = 'sbcdefghijk'
print(s.find('f'))
print(s.find('f',2,7))
print(s.find('a',2,7))
```

 2.index()

```python
s = 'sbcdefghijk'
print(s.index('c'))
print(s.index('z'))
```

```python
#Position of substring in the given main str
st = 'adakfjsabdz'
ss = 'a'
c = 0
for i in range(len(st)):
    if st[i] == ss:
        c+=1
        print('Found at index',i)
print('Total occurance: ',c)
```

```python
st = 'adakfjsabdz'
ss = 'a'
c = 0
p = 0
i = 0
while True:
    i = st.find(ss,p,len(st))
    if i == -1:
        break
    else:
        c+=1
        print('Found at index', i)
        p = i+len(ss)
print('The total no of occurance: ', c)
```

Changing Cases of str:
1. upper()
2. lower()
3. swapcase()
4. title() :- Capitalize all the words
5. capitalize() :- Only the first char will be uppercase

```python
s = 'learning Python is eaSy'
print(s.upper())
print(s.lower())
print(s.swapcase())
print(s.title())
print(s.capitalize())
```
