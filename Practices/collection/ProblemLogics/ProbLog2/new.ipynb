```python
def ab(n, o = 0):
    while n>0:
        r = n%10
        o+=r
        n//=10
    print(o)
ab(145)
```

```python
#Anagram
a = 'silent'.lower()
b = 'listen'.lower()
c = 0
if len(a) == len(b):
    for i in a:
        if a.count(i) == b.count(i):
            c+=1
    if c ==  len(a):
        print('Anagram')
    else:
        print('Not Anagram')
else:
    print('Not Anagram')
```

```python
#contineous order
l = [1, 3, 5, 6, 8]
n = l[1]-l[0]
print(n)
op = []
for i in range(len(l)-1):
    if l[i+1]-l[i]==n:
        op.extend([l[i],l[i+1]])
print(op)
```
