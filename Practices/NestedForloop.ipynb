```python
ip = 'bacbcaabbaa'
# op  = 'b4a5c2'
op = ''
for i in ip:
    if i not in op:
        c = 0
        for j in ip:
            if i == j:
                c+=1
        op += i+str(c)
print(op)
```

```python
l = [100, 200, 50, 400, 300]
n = 300
# out = [[100,200], [300]]
out = []
for i in range(len(l)):
    if l[i] == n:
        out.append([l[i]])
    for j in range(i+1, len(l)):
        if l[i]+l[j] == n:
            out.append([l[i],l[j]])
print(out)
```

```python
# No is Strong or not 145
n = 145
sum = 0
for i in str(n):
    fact = 1
    for j in range(int(i),0,-1):
        fact *= j
    sum+=fact
print(sum)
if sum == n:
    print("It's a strong no.")
```

```python
# Strong no without converting it to string
n = 145
temp = n
s = 0
for i in range(len(str(temp))):
    r = temp%10
    f = 1
    temp //= 10
    # print(r)
    for j in range(r,0,-1):
        f*=j
    s += f
if s == n:
    print('Strong')
```
