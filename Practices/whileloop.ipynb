```python
#Perfect num or not
n =int(input("Enter num: "))
i =  1
sum = 0
while i<n:
    if n%i==0:
        sum+=i
    i+=1
if sum == n:
    print('It is a perfect number.')
else:
    print('Not a perfect number.')
```

```python
#Those values which are in odd index even nums
t = eval(input('Enter tuple: '))
i = 0
l = []
if type(t) == tuple:
    while i < len(t):
        if type(t[i]) == int:
            if i%2 != 0:
                if t[i]%2==0:
                   l.append(t[i])
        i+=1
print(tuple(l))
```

```python
#Remove duplicates from a li
l = eval(input('Enter List:'))
i = 0
el = []
if type(l) == list:
    while i < len(l):
        if l[i] not in el:
            el.append(l[i])
        i+=1
print(el)
```

```python
#Sum of odd no. in given range
start = int(input('Enter start index: '))
end = int(input('Enter end index: '))
s = start+1
sum = 0
while s < end:
    sum += s
    s+=1
print(sum)
```

```python
#Gretest of a list
l = [1,2,3,4]
i = 0
g = l[0]
while i < len(l):
    if g < l[i]:
        g = l[i]
    i+=1
print(g)
```

```python
# Smallest bet a list
l = [1,2,3,4]
m = l[0]
i = 0
while i < len(l):
    if m > l[i]:
        m = l[i]
    i+=1
print(m)
```

```python
#Reverse the string without slicing
s = input('Enter string: ')
rev = ''
i = 0
while i < len(s):
    rev = s[i]+rev
    i+=1
print(rev)
```

```python
#Reverse a int
d = int(input('Enter num: '))
rev = 0
while d>0:
    r = d%10
    rev = rev*10+r
    d = d//10
print(rev)
```

```python
#Sum of each digit
n = int(input('Enter no: '))
sum = 0
while n > 0:
    r = n%10
    sum+=r
    n//=10
print(sum)
```

```python
#Product of digits
n = int(input('Enter no: '))
pro = 1
while n>0:
    r = n%10
    pro = pro*r
    n//=10
print(pro)
```

```python
#Fact of a num
n = int(input('Enter no: '))
fact = 1
while n>0:
    fact *= n
    n-=1
print(fact)
```

```python
#Toggle the string
s = input('Enter string: ')
u = ''
i = 0
while i<len(s):
    if 'a'<=s[i]<='z':
       u = u+chr(ord(s[i])-32)
    elif 'A'<=s[i]<='Z':
       u = u+chr(ord(s[i])+32)
    else:
        u = u+s[i]
    i+=1
print(u)
```

```python
#Find the sum of cube of a number in a str
s = input('Enter num: ')
sum = 0
i = 0
while i<len(s):
    if '0'<=s[i]<='9':
        sum += int(s[i])**3
    i+=1
print(sum)
```

```python
#Kth Rotation
k = int(input('Enter k: '))
s = [1,2,3]
i = 0
while i<k:
    s.insert(0,s[-1])
    s.pop()
    print(s)
    i+=1
```

```python
#Reversed Kth Rotation
k = int(input('Enter k: '))
s = [1,2,3]
i = 0
while i<k:
    s.append(s[0])
    s.pop(0)
    print(s)
    i+=1
```
