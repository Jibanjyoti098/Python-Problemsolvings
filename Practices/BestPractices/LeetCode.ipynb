Q1. Two Sums

```python
# ip = [2,7,11,15]
# target = 9
# op = [0,1]
num = eval(input('Enter sequence: '))
op = []
target = int(input('Enter Target: '))
for i in range(len(num)-1):
    if num[i] + num[i+1] == target:
        op.extend([i, i+1])
print(op)
```

```python
# nums = [2,7,11,15]
# target = 9
def twosum(nums, target):
    num_dict = {}
    for i, num in enumerate(nums):
        complement = target - num

        if complement in num_dict:
            return [num_dict[complement],i]

        num_dict[num] = i
twosum([2,7,11,15], 9)
```

Q2. Add two numbers (linked list)

```python
# l1 = [2,4,3]
# l2 = [5,6,4]
# op = [7,0,8]
# explanation : 342+465 = 807
# Using basic logic Time complexity: O(n)

l1 = [2,4,3]
l2 = [5,6,4]
op = []
s1 = ''
s2 = ''
for i in range(len(l1)):
    s1+=str(l1[i])
    s2+=str(l2[i])
sum = str(int(s1[::-1])+int(s2[::-1]))[::-1]
for i in range(len(sum)):
    op.append(int(sum[i]))
print(op)
```
