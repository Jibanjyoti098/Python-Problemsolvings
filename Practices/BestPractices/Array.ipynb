Q3. Reverse an array without using built-in functions

```python
ip = [1,2,3,4]
na = []
print(ip[::-1])
for i in range(1,len(ip)+1):
    na.append(ip[-i])
print(na)
```

Q4. Given an array of integers, count how many elements are even.

Input:
[1, 2, 3, 4, 5, 6]

Output:
3

```python
Input = [1, 2, 3, 4, 5, 6]
c = 0
for i in Input:
    if i%2==0:
        c+=1
print(c)
```

Q5.Two Sum (VERY IMPORTANT 🔥)

Problem:
Find two indices such that their values add up to target.

Input:
nums = [2, 7, 11, 15], target = 9
Output: [0, 1]

```python
nums = [2, 7, 11, 15]
target = 9
out = []
for i in nums:
    for j in nums:
        if i+j==target:
            if i not in out or i not in out:
                out.extend([i,j])
print(out)
```

Q6.Best Time to Buy and Sell Stock

Input: [7,1,5,3,6,4]
Output: 5

```python
inp = [7,1,5,3,6,4]
m = 7
d = 0
for i in range (len(inp)):
    for j in range (i+1,len(inp)):
        if i>j:
            m = j
        if d<j-i:
            d = j-i
print(d)
```

```python
price = [7,1,5,3,6,4]
min = price[0]
profit = price[1]-price[0]
for i in price:
    if min > i:
        min = i
    else:
        if i-min > profit:
            profit = i-min
print(profit)
```

Q8. Product of Array Except Self ⚡

* Input: [1,2,3,4]
* Output: [24,12,8,6]

```python
a = [1,2,3,4]
o = []
for i in a:
    m = 1
    for j in a:
        if i!=j:
            m*=j
    o.append(m)
print(o)
```
