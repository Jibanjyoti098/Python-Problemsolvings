Polymorphism in function

```python
from numpy.ma.core import power


def mul(n,m):
    return n*m

print(mul(2,'h'))
```

Area = pi*r**2, circumference = 2*pi*r

```python
def circle(pi, r):
    area = pi*(r**2)
    circum = 2*pi*r
    return  area, circum

print(circle(3.14, 2))
```

Default parameter value

```python
def greet(name='jiban'):
    print('Greetings {}'.format(name))
greet('risi')
```

Lambda function

```python
cube = lambda x: x**3
print(cube(3))
```

function with *args

```python
def sumdig(*args):
    return sum(args)
print(sumdig(1,2,3,4))
```

function with **kwargs

```python
def print_para(**kwargs):
    for i,j in kwargs.items():
        print(f'{i}:{j}')

print_para(name='risi', power='darkness', enemy='light')
```

Generator function with yeild

```python
def even(limit):
    for i in range(2,limit+1,2):
        yield i
for n in even(10):
    print(n)
```

Recursive function

```python
#Factorial
def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n-1)
print(fact(4))
```
