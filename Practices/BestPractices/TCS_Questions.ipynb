Q2. Maximum Sum of Elements Less Than or Equal to Given Limit
Problem Statement
You are given:Scripting Languages

1. An array of integers of size n
2. An integer maxSum (maximum allowed sum)

```python
n = int(input('Enter n: '))
arr = eval(input('Enter array'))#2,3,5,7
maxSum = int(input('Enter n: '))
m = arr[0]
r = 0
for i in arr:
    res = 0
    for j in arr:
        if i+j==maxSum:
            res+=i+j
        else:
            if m<i:
                m = i
                res = m
    r = res
print(r)
```
