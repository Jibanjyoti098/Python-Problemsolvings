Inheritance:

```python
class Parent:
    a = 'apple'
    b = 'bat'
class Child(Parent):
    @classmethod
    def show(cls):
        print(cls.a,cls.b)
c = Child()
c.show()
Child.a
```

Constructor Chaining:

```python
class Bank:
    def __init__(self, n, a):
        self.n = n
        self.a = a
    def show(self):
        print(self.n, self.a)
class SBI:
    def __init__(self):
        super().__init__(n, a)
    def show2(self):
        super().show()
o =  Bank('Jiban', 22)
o.show()
```

Multilevel Inheritance:

```python
class Ten:
    cl = 10
    school = 'ssvm'
    secured = 61
    @classmethod
    def show(cls):
        print(cls.cl, cls.school, cls.secured)
class Twelve(Ten):
    cl2 = 12
    school2 = 'sdcs'
    secured2 = 68
    @classmethod
    def show2(cls):
        super().show()
        print(cls.cl2, cls.school2, cls.secured2)
class Deg(Twelve):
    cl3 = 'bsc'
    school3 = 'ln'
    secured3 = 64
    @classmethod
    def show3(cls):
        super().show2()
        print(cls.cl3, cls.school3, cls.secured3)
obj = Deg()
obj.show3()
```
