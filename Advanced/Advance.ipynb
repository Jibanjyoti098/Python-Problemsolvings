Lambda Function:-

```python
even = lambda n: n%2==0
print(even(3))
```

to add min 2 or max 5 nums

```python
x = lambda m,n,o=0,p=0,q=0: m+n+o+p+q
x(2,3,4)
```

if its start with vowel if it is then return whole string in uppercase else return the reversed string

```python
x  = lambda s: s if s[0].lower() in 'aeiou' else s[::-1]
x(input())
```

If the list is palindrome if it is create a nested list that will have the same same value as the original list and store it into the original list else delete all the data from the list

```python
x = lambda l: [l] if l[::-1]==l else l.clear()
x([1,2,1])
```
